## Day 4 
**Lists & Tuples**

```python
# 1. Create a list of 5 numbers
l1 = [1,2,3,4,5,6]
print("l1 is: ",  l1)

# 2. Append a new number
l1.append(9)
print("l1 after append: ",l1)

# 3. Remove one element
l1.remove(4)
print("l1 after remove: ", l1)

# 4. Access first element
print("first element of list: ", l1[0])

# 5. Find list length
print("length of list is: ",(len(l1)))

# 6. Create a tuple of 3 items
T1 = (6,7,8)
print("tuple: ",T1)

# 7. Access tuple element
print("first element of tuple: ",(T1[0]))

# 8. Create list of names
l2 = ['faiza', 'amna', 'ayesha']
print('list of names: ', l2)

# 9. Change second item in list
l2[1] = 'umer'
print('New list after changing 2nd element is: ',l2)

# 10. Try changing tuple value (observe error)
T1[2] = 9

T2 = (3,4,5)
T2[2] = 9
print(T2)
```

## Day 4 
**Dictionaries & Sets**

```python
# 1. Create dictionary with name and age
D1 = {'name':'Hadi' ,'age' : 13}
print ('Dict: ',D1)

# 2. Access value by key
print('Access value by key:',D1['name'])

# 3. Add new key‑value pair
D1 ['gender'] = 'male'
print("adding new key value: ", D1)

# 4. Print all keys
print("# 4. Print all keys: ", D1.keys())

# 5. Print dictionary length
print("length of dictionary: ", len(D1))

# 6. Create set with duplicate values
s1 = {1,2,3,2,3,2,4,5}
print ('set: ', s1)

# 7. Add new value to set
s1.add(6)
print('new value to set: ', s1)

# 8. Remove value from set
s1.remove(2)
print ('new set after removal: ', s1)

# 9. Check membership in set
print  ('check membership  in set: ', 3 in s1)
print ('check membership  in set:', 9 in s1)

# 10. Convert list to set
l1 = [1, 2, 2, 3, 4, 4, 5]
s2 = set(l1)
print('list into set: ',s2)
```
