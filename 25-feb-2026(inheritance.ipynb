```python
class Employee:

    def __init__(self, name, employee_id, salary):
        # Private attributes
        self.__name = name
        self.__employee_id = employee_id
        self.__salary = salary
        self._department = "General"

    # Getters
    def get_name(self):
        return self.__name

    def get_employee_id(self):
        return self.__employee_id

    def get_salary(self):
        return self.__salary

    def get_department(self):
        return self._department

    # Setters
    def set_salary(self, new_salary):
        if new_salary > 0:
            self.__salary = new_salary
            return True
        else:
            print("Salary must be positive!")
            return False

    def set_department(self, new_department):
        self._department = new_department

    # Methods
    def calculate_bonus(self):
        return self.__salary * 0.05

    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"Employee ID: {self.__employee_id}")
        print(f"Department: {self._department}")
        print(f"Salary: ${self.__salary:,.2f}")
        print(f"Bonus: ${self.calculate_bonus():,.2f}")



class Manager(Employee):

    def __init__(self, name, employee_id, salary, team_size):
        super(). __init__(name, employee_id, salary)
        self.__team_size = team_size
        self.set_department("Management")

    def get_team_size(self):
        return self.__team_size

    def set_team_size(self, new_size):
        if new_size > 0:
            self.__team_size = new_size
            return True
        return False

    def display_info(self):
        super().display_info()
        print(f"Team Size: {self.__team_size} members")
        print(f"Management Bonus: +${self.__team_size * 500:,.2f}")


emp = Employee("Alice", "E123", 75000)
print(emp.calculate_bonus())
print(emp.get_name())

m = Manager("Bob", "M456", 90000, 5)
m.display_info()
```

```python
class Vehicle:
    def __init__(self, brand, speed, color, doors):
        self.__brand = brand      # private
        self._speed = speed       # protected
        self._color = color
        self._doors = doors

    def get_brand(self):
        return self.__brand

    def get_speed(self):
        return self._speed
    
    def get_color(self):
        return self._color
    
    def get_doors(self):
        return self._doors

    def display_vehicle(self):
        print(f"Brand: {self.__brand}")
        print(f"Speed: {self._speed} km/h")
        print(f"Color: {self._color}")
        print(f"Doors: {self._doors}")
        

class Car(Vehicle):
    def __init__(self, brand, speed, color, doors,model):
        super().__init__(brand, speed, color,doors)   # parent constructor
        self.__model = model             # private

    def get_doors(self):
        return self.__model

    # method overriding (polymorphism)
    def display_vehicle(self):
        super().display_vehicle()        # parent method call
        print(f"Model: {self.__model}")

c = Car("Sports car", 200, 'Red', 4, "1800CC")

print("Car Information:")
c.display_vehicle()
```

```python
class Animal:
    def __init__(self, name, color, legs, voice):
        self._name = name
        self._color = color
        self.__legs = legs
        self.__voice = voice

    def get_name(self):
        return self._name

    def get_color(self):
        return self._color

    def get_legs(self):
        return self.__legs

    def get_voice(self):
        return self.__voice

    def display_Animal(self):  
        print(f"Name: {self._name}")
        print(f"Color: {self._color}")
        print(f"Legs: {self.__legs}")
        print(f"Voice: {self.__voice}")


class Elephant(Animal):
    def __init__(self, name, color, legs, voice, ears):
        super().__init__(name, color, legs, voice)
        self.__ears = ears

    def get_ears(self):
        return self.__ears

    def display_Animal(self):
        super().display_Animal()   
        print(f"Ears: {self.__ears}")

e = Elephant("Elephant", "Grey", 4, "Trumpet", "Big")
print("Elephant Information:")
e.display_Animal()
```
