```python

arr1D = np.arange(1, 7)  
print("1D Array:", arr1D)

print("Indexing arr1D[2]:", arr1D[2])        
print("Slicing arr1D[2:7]:", arr1D[2:5])    

arr2D = arr1D.reshape(2, 3)   
print("\nConverted 2D Array:\n", arr2D)

print("Element at [1,1]:", arr2D[1, 1])   

arr2D[1, 1] = 99
print("After replacing element at [1,1]:\n", arr2D)

arr1D_large = np.arange(1, 7)   
arr3D = arr1D_large.reshape(1, 3, 2)  
print("Converted 3D Array (1x3x2):", arr3D)

print("Element at [2,4,0]:", arr3D[0, 2, 1])  

arr1D_back = arr3D.flatten()
print("Converted back to 1D:", arr1D_back)
```
